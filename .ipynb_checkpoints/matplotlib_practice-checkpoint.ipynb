```python
# Import dependencies.
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

### recreate the line chart we just made using the object-oriented interface method
1. The plt.subplots() function returns a tuple that contains a figure and axes object(s).
2. fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
3. The first variable, fig, references the figure.
4. The second variable, ax, references the axes of the figure and allows us to annotate the figure.
5. The ax.plot(x_axis, y_axis) plots the data from the x and y axes

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

### The fig, ax = plt.subplots() line of code can also be written like this:

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

### We can adjust the size by multiplying the data in the y-axis by any value. Let's multiply each data point in the y-axis by 3

```python
y_axis_new = []
y_axis_new = y_axis * 3
y_axis_new[:5]
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
y_axis_larger
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

### Using the list comprehension technique to multiply each data point in the y-axis list, we would do the following in the plt.scatter() function:

```python
plt.scatter(x_axis, y_axis, s= [ (i*3) for i in y_axis] )
```

```python
# The % before and after the .1f% formats the number as a percentage.

# Note: The plt.subplots should always come before the plt.pie statement.

plt.subplots(figsize=(6,6))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Changing the major ticks on the X axis
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Object oriented approach
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Adding Minor ticks

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
